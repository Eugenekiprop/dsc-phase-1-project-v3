```python
#display sum of missing values
df.isna().sum()
```

```python
#display dataframe after identifying missing values and dropping them
def handle_missing_values(df):

  missing_values = df.isnull().sum()
  missing_percentage = (missing_values / len(df)) * 100
  missing_info_df = pd.DataFrame({
      'Missing Count': missing_values,
      'Missing Percentage': missing_percentage
  })
  missing_info_df = missing_info_df[missing_info_df['Missing Count'] > 0].sort_values(by='Missing Count', ascending=False)

  df_dropped = df.dropna()

  return missing_info_df, df_dropped

# Example usage with the 'df' DataFrame from the preceding code:
missing_data_info, df_no_missing = handle_missing_values(df)

print("Missing value information:")
print(missing_data_info)

print("\nDataFrame after dropping rows with missing values:")
print(df_no_missing.head())
print("\nShape of DataFrame after dropping missing values:", df_no_missing.shape)
```

OUTLIERS AND REMOVING THE OUTLIERS

```python
#define a function for visualizing outliers
def visualize_outliers(df, column):
  plt.figure(figsize=(10, 6))
  sns.boxplot(x=df[column])
  plt.title(f'Box plot for {column} to visualize outliers')
  plt.xlabel(column)
  plt.show()

# Example usage for a numerical column like 'total_fatal_injuries'
if 'total_fatal_injuries' in df.columns:
  visualize_outliers(df, 'total_fatal_injuries')

# Example usage for a numerical column like 'total_fatal_injuries'
if 'total_uninjured' in df.columns:
    visualize_outliers(df, 'total_uninjured')

```

```python
#define a function for removing outliers
def remove_outliers_iqr(df, column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  df_cleaned = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)].copy()
  return df_cleaned

# Remove outliers from 'Age'
if 'total_fatal_injuries' in df.columns:
    df_cleaned_age = remove_outliers_iqr(df, 'total_fatal_injuries')
    print(f"Shape before removing Age outliers: {df.shape}")
    print(f"Shape after removing Age outliers: {df_cleaned_age.shape}")
    visualize_outliers(df_cleaned_age, 'total_fatal_injuries')
    df = df_cleaned_age # Update df to the cleaned version

# Remove outliers from 'Fare'
if 'total_uninjured' in df.columns:
    df_cleaned_fare = remove_outliers_iqr(df, 'total_uninjured')
    print(f"Shape before removing Fare outliers: {df.shape}")
    print(f"Shape after removing Fare outliers: {df_cleaned_fare.shape}")
    visualize_outliers(df_cleaned_fare, 'total_uninjured')
    df = df_cleaned_fare # Update df to the cleaned version

print("\nFinal DataFrame shape after removing outliers from Age and Fare:", df.shape)
```
